```python
draw_script = """\

function draw_array2d(where, array, opts) {
  var block_size = opts.block_size,
      margin = opts.margin;
  var outer = 1. * (block_size + margin);
    
  var rows = where.selectAll('g')
                  .data(array).enter()
                  .append('g')
                  .attr('kind', 'row')
                  .attr('transform', 
                        function (d, i) {
                         return ('translate(0,' + 
                           (i * outer) + ')');});

  var cells = rows.selectAll('g') // rect and text
                  .data(function (d) {return d;})
                  .enter().append('g')
                  .attr('transform', 
                        function (d, i) {
                          return ('translate(' + 
                                  ((i+1) * outer) + ', 0)');});

  cells.append('rect')
       .style('fill', 'white')
       .style('stroke', "black")
       .attr('width', block_size)
       .attr('height', block_size);

  cells.append('text')
       .attr('x', 0)
       .attr('y', outer / 2)
       .text(function (d) {return d;});

  var text_width, x,
      text = cells.selectAll('text');
  for (var i=0; i<text.length; i++) {
      text_width = text[i][0].getBBox().width;
      x = 1. * (block_size - text_width) / 2;
      d3.select(text[i][0]).attr('x', x);
  }
};

function draw_sort_index(where, argsort, opts) {    
  var outer = opts.block_size + opts.margin;
  var rows = where.selectAll('[kind=row]')[0];
  console.log(rows);
  
  for (var i=0; i<argsort.length; i++) {
    (function (i) {  
      d3.select(rows[argsort[i] + 1]).transition().duration(opts.duration)
      .attr('transform', function (d) {
        return ('translate(0,' + 
                ((i + 1) * outer) + ')');
      });
    }(i));
  }
};

"""
```
